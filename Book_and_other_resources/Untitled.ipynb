```python
def write_gene_list(gencode_gtf):
    gene_list=[]
    i=0
    with open(gencode_gtf) as gtf, open("gene_list.txt", 'w') as out:
        for line in gtf:
            i+=1
            if not line.startswith("#"):
                lista = line.strip().split("\t")
                if len(lista) >= 9:
                    if lista[2].strip() == "gene":
                        chromosome = lista[0]
                        gene_name = lista[8].split(";")[0].split()[1].split('"')[1]
                            
                        if not (chromosome, gene_name) in gene_list:
                            gene_list.append((chromosome, gene_name))
                            out.write("{}\t{}\n".format(chromosome, gene_name))
                        else:
                            pass
                else:
                    pass
    return gene_list
```

```python
def load_gene_list(gene_list_txt):
    
    gene_list=[]
    
    with open(gene_list_txt) as genes:
        for line in genes:
            lista=line.strip().split()
            gene_list.append((lista[0],lista[1]))
    
    return gene_list
            
```
